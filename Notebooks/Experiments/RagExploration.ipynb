# Guitar Alchemist: RAG & OPTIC-K Exploration

This notebook demonstrates how to experiment with the Guitar Alchemist RAG system directly using C#.
Using a C# notebook allows you to step into the `MusicalEmbeddingGenerator` logic and visualize vector partitions.

```
// 1. Reference the Local Projects
#r "project:../../Common/GA.Domain.Core/GA.Domain.Core.csproj"
#r "project:../../Common/GA.Business.ML/GA.Business.ML.csproj"

using GA.Domain.Core.Notes;
using GA.Domain.Core.Intervals;
using GA.Domain.Core.Tonal;
using GA.Business.ML.Embeddings;
using GA.Business.ML.Embeddings.Services;
using GA.Business.ML.Rag.Models;
using Microsoft.DotNet.Interactive.Formatting;
```

## 2. Initialize the Embedding Generator
We instantiate all component services required by the OPTIC-K schema.

```
// Helper factory to create the generator with all dependencies
MusicalEmbeddingGenerator CreateGenerator() => new MusicalEmbeddingGenerator(
    new IdentityVectorService(),
    new TheoryVectorService(),
    new MorphologyVectorService(),
    new ContextVectorService(),
    new SymbolicVectorService(),
    new ModalVectorService(),
    new PhaseSphereService()
);

var generator = CreateGenerator();
Console.WriteLine("Generator Initialized Successfully.");
```

## 3. Generate an Embedding
Let's create a chord (e.g., C Major 7) and generate its vector representation.

```
// Define a Cmaj7 chord (C, E, G, B)
var cMaj7Notes = new[] { 0, 4, 7, 11 }; // Midi Note Numbers (C4, E4, G4, B4 usually, here just PC-relative)

// Create a mock document
var cMaj7 = new ChordVoicingRagDocument 
{
    Id = Guid.NewGuid().ToString(),
    ChordName = "Cmaj7",
    Notes = ["C", "E", "G", "B"],
    PitchClasses = [0, 4, 7, 11],
    MidiNotes = [60, 64, 67, 71], // C4, E4, G4, B4
    Root = "C",
    RootPitchClass = 0,
    Quality = "Major7",
    Bass = "C",
    MidiBassNote = 60,
    IntervalClassVector = [0,0,1,1,1,0], // Example ICV
    Consonance = 0.8,
    Brightness = 0.6,
    SemanticTags = ["jazz", "open"]
};

// Generate Embedding
var embedding = await generator.GenerateEmbeddingAsync(cMaj7);

Console.WriteLine($"Embedding Dimension: {embedding.Length}"); // Should be 216 for OPTIC-K
Display(embedding.Take(20)); // Show first 20 dimensions
```
