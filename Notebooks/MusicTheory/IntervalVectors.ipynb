# Interval Class Vectors Navigation

This notebook demonstrates navigation of musical objects using the Guitar Alchemist Core library.
It lists all possible **Interval Class Vectors** (ICV) associated with their Forte Numbers and Prime Forms.

```
#r "/app/GA.Business.Core.dll"
#r "/app/GA.Core.dll"
using GA.Business.Core.Atonal;
using System.Text;
using System.Linq;
using Microsoft.DotNet.Interactive;

// Pre-warm the catalog to ensure data is loaded
ProgrammaticForteCatalog.PreWarm();

var sb = new StringBuilder();
sb.Append("<div style='height:600px; overflow-y:auto;'>");
sb.Append("<table class='table table-striped table-hover'>");
sb.Append("<thead style='position:sticky; top:0; background:#222;'><tr><th>Forte #</th><th>Prime Form</th><th>Interval Class Vector</th><th>Cardinality</th></tr></thead>");
sb.Append("<tbody>");

var catalog = ProgrammaticForteCatalog.ForteByPrimeFormId;
var setClasses = SetClass.Items
    .Where(sc => catalog.ContainsKey(sc.PrimeForm.Id))
    .Select(sc => new { SetClass = sc, Forte = catalog[sc.PrimeForm.Id] })
    .OrderBy(x => x.Forte)
    .ToList();

foreach(var item in setClasses)
{
    var f = item.Forte;
    var sc = item.SetClass;
    var icv = sc.IntervalClassVector;
    
    // Format ICV like <254361>
    var icvStr = string.Join("", icv.Vector.Values);

    sb.Append($"<tr>");
    sb.Append($"<td><strong>{f}</strong></td>");
    sb.Append($"<td>{sc.PrimeForm}</td>");
    sb.Append($"<td><code>&lt;{icvStr}&gt;</code></td>");
    sb.Append($"<td>{sc.Cardinality.Value}</td>");
    sb.Append($"</tr>");
}

sb.Append("</tbody></table></div>");

// Display as HTML
Display(sb.ToString(), "text/html");
```
